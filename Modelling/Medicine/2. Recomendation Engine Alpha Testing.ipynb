```python
import pandas as pd
```

```python
df = pd.read_csv('final_test.csv')
df.head()
```

```python
df = df.groupby(['condition','drugName']).agg({'total_pred' : ['mean']})
```

```python
df.columns
```

```python
df.loc['ADHD'].sort_values([('total_pred', 'mean')], ascending=False)[:5]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def medication(condition):
    medicines = df.loc[condition].sort_values([('total_pred', 'mean')], ascending=False)[:5]
    medicines = medicines.reset_index()
    #print(medicines.columns)
    plt.figure(figsize = (10,7))
    fig = sns.barplot(x = ('drugName',''), y = ('total_pred', 'mean'), data=medicines)
    plt.xlabel('Top Medicines', fontsize=15)
    plt.ylabel('Effectiveness and Goodness', fontsize=15)
    return fig
    
```

```python
for i in range(10):
    condition = str(input('Enter the condition'))
    img = medication(condition)
    plt.show()
```
